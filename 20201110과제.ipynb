# 연습문제

```python
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic.age = titanic.age.fillna(titanic.age.mean()) 
#데이터 사이언스 스쿨의 앞에 연습문제에서 age가 NaN인경우 전부 평균값나이로 올렸습니다
#그래서 청년,중년이 더 많을 수 있습니다.
# 타이타닉호의 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 고객의 평균 나이 값이 되도록 titanic 데이터프레임을 고쳐라.

```

### 4.4.7 각 나이그룹 나누고 비율 그리고 비율전체합 1

```python
#그룹으로 나누기
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
titanic["age_cat"] = pd.cut(titanic.age, bins, labels=labels)
titanic["age_cat"]
```

```python
titanic.age_cat.value_counts()
```

```python
# 비율
k = titanic.age_cat.value_counts()/titanic["age_cat"].count()
print(k)
```

```python
# 비율 합이 1이 나와야 함
k.sum()
```

### 4.4.8 카테고리3

```python
titanic["category3"] = titanic.apply(lambda r:"미성연자" if r["age"] < 20 else f'{r["age_cat"]}{"남성"if r.sex == "male" else "여성"}',axis=1)
titanic["category3"]
```
