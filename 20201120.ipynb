```python
lat = []
lng = []
name = []
locate = []
for i in df.index:
    if df["도로명주소"][i].split()[0] == "서울특별시" and df["도로명주소"][i].split()[1] == "강서구":
        # addr = f'{df["도로명주소"][i]}'
        # url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + addr
        # result = requests.get(urlparse(url).geturl(),headers={"Authorization":"KakaoAK "+kakao_key}).json()
        lat.append(df['경도'][i])
        lng.append(df["위도"][i])
        name.append(df["상호명"][i])
        locate.append(df["지점명"][i] if df["지점명"][i] != 'nan' else '')
a = pd.DataFrame({
                '상호명':name,
                "지점명":locate,
                '위도':lat,
                '경도':lng
})
a
```

```python
ediya = a[a["상호명"].str.contains("이디야")]
ediya["c_code"] = "#3388ff"
ediya
# def find_cafe(a):
#     for i in a.index:
#         if a["상호명"][i].find("스타벅스") >= 0:  
#             print(a[i:i+1]) 
# find_cafe(a)
```

```python
paik = a[a["상호명"].str.contains("빽다방")]
paik["c_code"] = "springgreen"
paik
```

```python
coffee_bean = a[a["상호명"].str.contains("커피빈")]
coffee_bean["c_code"] = "red"
coffee_bean
```

```python
twosome = a[a["상호명"].str.contains("투썸플레이스|tw")]
twosome["c_code"] = "blue"
twosome
```

```python
starbucks = a[a["상호명"].str.contains("스타벅스|star")]
starbucks["c_code"] = "green"
starbucks
```

```python
mapping = folium.Map(location=[starbucks["경도"].mean(),starbucks["위도"].mean()],zoom_start=11)
for i in starbucks.index:
    folium.Marker(
                [starbucks["경도"][i],starbucks["위도"][i]], 
                popup=f'starbucks,{starbucks["지점명"][i]}').add_to(mapping)
for i in twosome.index:
    folium.Marker(
                [twosome["경도"][i],twosome["위도"][i]],
                color = "r", 
                popup=f'twosome, {twosome["지점명"][i]}',
                icon=folium.Icon(color = 'red')).add_to(mapping)
for i in paik.index:
    folium.Marker(
                [paik["경도"][i],paik["위도"][i]],
                color = "r", 
                popup=f'paik, {paik["지점명"][i]}',
                icon=folium.Icon(color = 'green')).add_to(mapping)
# mapping
```

```python
mapping = folium.Map(location=[paik["경도"].mean(),paik["위도"].mean()],zoom_start=15)
cafes = [starbucks,twosome,paik,coffee_bean,ediya]
for cafe in cafes:
    for i in cafe.index:
        folium.Marker(
                    [cafe["경도"][i],cafe["위도"][i]],
                    color = "r", 
                    popup=f'{cafe["상호명"][i]}, {cafe["지점명"][i]}',
                    icon=folium.Icon(color=cafe.c_code[i])).add_to(mapping)
    
# for i in starbucks.index:
#     folium.Marker(
#                 [starbucks["위도"][i],starbucks["경도"][i]], 
#                 popup=f'starbucks,{starbucks["지점명"][i]}').add_to(mapping)
# for i in twosome.index:
#     folium.Marker(
#                 [twosome["위도"][i],twosome["경도"][i]],
#                 color = "r", 
#                 popup=f'twosome, {twosome["지점명"][i]}',
#                 icon=folium.Icon(color = 'red')).add_to(mapping)
# for i in paik.index:
#     folium.Marker(
#                 [paik["위도"][i],paik["경도"][i]],
#                 color = "r", 
#                 popup=f'paik, {paik["지점명"][i]}',
#                 icon=folium.Icon(color = 'green')).add_to(mapping)
mapping
```
