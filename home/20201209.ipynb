```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import platform
import matplotlib as mpl 
from konlpy.tag import Okt,Kkma,Hannanum
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import nltk
from tqdm import tqdm_notebook
import time
import math
%matplotlib inline
# 한글폰트 사용
mpl.rc('font', family='Malgun Gothic')
mpl.rc('axes', unicode_minus=False)
```

```python
meeting_text = ''
answer_count = 0
page_num = 10
kin_url = 'https://kin.naver.com/search/list.nhn'
keyword = '생신선물'
driver = webdriver.Chrome('D:/workspace/data-home/05.Crawling/chromedriver.exe')
# 전체페이지
for n in tqdm_notebook(range(1,page_num+1)):
    url = f'{kin_url}?query={keyword}&page={n}'
    driver.get(url)
    # 질문들 하나씩 들어가기
    for k in driver.find_elements_by_css_selector('.basic1 > li > dl > dt'):
        k.click()
        driver.switch_to_window(driver.window_handles[1])
        answerCount = int(driver.find_element_by_css_selector('._answerCount.num').text)
        click_count = math.floor((answerCount-1)/5)
        # 더보기 버튼 누르기(5개씩 나옴)
        for a in range(click_count):
            driver.find_element_by_css_selector("#nextPageButton").click()
            time.sleep(1)
        answers = driver.find_elements_by_class_name('_endContentsText.c-heading-answer__content-user')
        # 답변 가져오기
        for answer in answers:
            meeting_text += '\n' + answer.text
            answer_count += 1
        driver.close()
        driver.switch_to_window(driver.window_handles[0])
driver.close()
answer_count
```

```python
with open('./data/meeting_text.txt','w',encoding='UTF-8') as f:
    f.write(meeting_text)
```

```python
with open('./data/meeting_text.txt','r',encoding='UTF-8') as f:
    meeting_text = f.read()
```

```python
stop_words = ['상견례','장소','곳','서울','추천','음식','것','요리','분위기','회원','저','전',]
```

```python
okt = Okt()
meeting = [word for word in okt.nouns(meeting_text) if word not in stop_words]
ko = nltk.Text(meeting,name='상견례 장소')
ko.vocab().most_common()
```
