```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import platform
import matplotlib as mpl 
from konlpy.tag import Okt,Kkma,Hannanum
%matplotlib inline
# 한글폰트 사용
mpl.rc('font', family='Malgun Gothic')
mpl.rc('axes', unicode_minus=False)
```

```python
import nltk
```

```python
from konlpy.corpus import kobill

files_ko = kobill.fileids()
doc_ko = kobill.open('1809890.txt').read()
```

```python
doc_ko[:1000]
```

```python
okt = Okt()
tokens_ko = okt.nouns(doc_ko)
tokens_ko[:10]
```

```python
ko = nltk.Text(tokens_ko, name = '대한민국 국회 의안 제 1809890호')
len(ko.tokens), len(set(ko.tokens))
```

```python
ko.vocab()
```

```python
plt.figure(figsize = (12, 6))
ko.plot(50) # 사용빈도가 높은 단어 50개에 대해서 plot 그리기
plt.show()
```

```python
stop_words = ['.', '(', ')', ',', "'", '%', '-', 'X', ').', '×',
              '의', '자', '에', '안', '번', '호', '을', '이', '다', '만', '로', '가', '를']
ko = [each_word for each_word in ko if each_word not in stop_words]
ko[:10]
```

```python
ko = nltk.Text(ko, name = '대한민국 국회 의안 제 1809890호')

plt.figure(figsize = (12, 6))
ko.plot(50)
plt.show()
```

```python
ko.count('초등학교')
```

```python
# 원하는 단어의 문서 내 개략적인 위치와 분량 확인
plt.figure(figsize = (12, 6))
ko.dispersion_plot(['육아휴직', '초등학교', '공무원'])
```

```python
ko.concordance('초등학교')
```

```python
ko.collocations()
```

```python
data = ko.vocab().most_common(150) # 출현 횟수가 높은 상위 150개의 단어들

wordcloud = WordCloud(font_path = "c:/Windows/Fonts/malgun.ttf",
                      relative_scaling = 0.2,
                      background_color = 'white',
                     ).generate_from_frequencies(dict(data))

plt.figure(figsize = (12, 8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
