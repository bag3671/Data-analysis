```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
mpl.rc('font', family='NanumGothic')
mpl.rc('axes', unicode_minus=False)
```

```python
cafe =  pd.read_csv("cafe_2year.csv")
cafe.head()
cafe.count()
```

```python
#1번
cafe[2::3].plot.bar(x="기준월",y="카페수")
plt.xlabel("년월")
plt.ylabel("카페수", rotation=0)
plt.ylim(min(cafe["카페수"]),max(cafe["카페수"]))
plt.title("전국의 월별 까페수")
plt.show()
```

```python
#2번
doc2 = pd.read_csv("doctor_2019.csv")
doc2
```

```python
doc3 = doc2.groupby("지역").sum().loc[['서울', '부산', '대구', '인천', '대전', '광주', '울산']]
doc3
```

```python
#3번
doc3.to_csv("doctor2.csv")
print("doctor2.csv 파일 쓰기 완료!")
```

```python
#4번
doc3 = pd.read_csv("doctor2.csv")
doc3.plot.scatter("지역","전문의 수")
plt.show()
```

```python
# 5번
doc3 = pd.read_csv("doctor2.csv") 
plt.plot(doc3["지역"],doc3['일반의 수'],'r.-',label='일반의')
plt.plot(doc3["지역"],doc3['인턴 수'],'s-',label='인턴')
plt.plot(doc3["지역"],doc3['레지던트 수'],'o-',label='레지던트')
plt.plot(doc3["지역"],doc3['전문의 수'],'--',label='전문의')
plt.ylabel("의사수")
plt.title("서울 및 광역시 종합병원 의사 수")
plt.legend()
plt.show()
```

```python
# 6번
doc3 = pd.read_csv("doctor2.csv")
doc3 = doc3.set_index("지역")
doc3 = doc3.rename(columns={'일반의 수':'일반의','인턴 수':'인턴','레지던트 수':'레지던트','전문의 수':'전문의'})
doc3.loc["서울"].plot.pie(autopct='%.1f%%',startangle=90)
plt.axis('equal')
plt.title("서울 종합병원 의사 유형별 분포")
plt.show()
```
