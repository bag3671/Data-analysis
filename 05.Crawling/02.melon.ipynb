```python
from bs4 import BeautifulSoup
import urllib.request
import urllib.parse
import pandas as pd
import requests
```

```python
url = 'https://www.melon.com/chart/week/index.htm'
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko'}
req = requests.get(url, headers = header)
html = req.text
soup = BeautifulSoup(html,'html.parser')
# web_url = "https://www.melon.com/chart/week/index.htm"
# with urllib.request.urlopen(web_url) as response:
#     html = response.read()
#     soup = BeautifulSoup(html, 'html.parser')
```

```python
title = soup.find_all("div", {"class": "ellipsis rank01"}) 
singer = soup.find_all("div", {"class": "ellipsis rank02"}) 
album = soup.find_all("div",{"class": "ellipsis rank03"})
diff = soup.find_all("span",{"class" : "rank_wrap"})
rank = soup.find_all("span",{"class": "rank"})[1:]
titles=[]; singers=[]; albums=[]; diffs = []; ranks = []
def append_s(x,y):
    for i in x:
        y.append(i.get_text().replace("\n",""))
append_s(title,titles)
for i in singer:
    k = len(i.get_text())
    singers.append(i.get_text()[0:k//2].replace("\n",""))
append_s(album,albums)
append_s(rank,ranks)
for i in diff:
    diffs.append(i.get_text().replace("\n"," "))
diffs
    # print(i.get_text().replace("\n","")[:5]) 
    # print(i.get_text().replace("\n","")[5:])
k = pd.DataFrame({
                'title':titles,
                'singer':singers,
                'album':albums,
                'diff':diffs,
                'rank':ranks})
k = k.set_index('rank')
k.head(20)
```
